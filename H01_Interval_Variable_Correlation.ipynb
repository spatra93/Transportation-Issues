```python
def get_redundant_pairs(d2):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = d2.columns
    for i in range(0, d2.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(d1):
    au_corr = d1.corr(method='pearson').abs().unstack()
    labels_to_drop = get_redundant_pairs(d1)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    au_corr = au_corr[au_corr > 0.65]
    return au_corr


print(get_top_abs_correlations(df_interval))
```
